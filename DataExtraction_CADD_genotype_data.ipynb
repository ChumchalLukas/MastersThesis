```python
## Used modules

import pandas as pd
import numpy as np

```

```python


#Filtered data with all GTsample 
data_exones = pd.read_excel('Exones_variace_GT.xlsx')

empty_exones = data_exones['EMPTY'] == 0
data_exones_1 = data_exones.loc[empty_exones,:]

#Filtered data with CADD_phred dot data
dot_contain_exones = data_exones_1['CADD_phred'] == '.'
dot_notcontain_exones = data_exones_1['CADD_phred'] != '.'

data_exones_1_dot = data_exones_1.loc[dot_contain_exones,:]
data_exones_1_nodot = data_exones_1.loc[dot_notcontain_exones,:]

# data_exones_1_dot.to_excel('VarExons_genotypes_CTD_dot_2022_1012.xlsx', index = False) #export data with only dots as CADD_phred! | not important
# data_exones_1_nodot.to_excel('pgvarOstE_nodot.xlsx') #data without CADD_phred == '.'

#Function to filter CADD_according threshold

def greater(x, threshold = 15.0):
    '''
    Checks if there is a value > theshold
    
    x values
     - comma separated string
     - dloat
     - empty
    
    '''
    if x and isinstance(x, str) and "," in x:
        # comma separated string
        # Check if list has a value > threshold
        return any (v > threshold for v in [float(i) for i in x.split(',')])
    elif x and isinstance(x, float):
        # Float
        return x > threshold
    else:
        #Empty
        return False
    

#Filter out rare-common and rare variants   
data_exones_2 = data_exones_1_nodot[data_exones_1_nodot.CADD_phred.apply(lambda x: greater(x, 15.0))]

#Exporting final file - výsledek
data_exones_2.to_excel('Exones_variance_pdfilter.xlsx', index = False)
```
